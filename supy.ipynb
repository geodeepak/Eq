## The SUSTREAM, SUFILE, SUBUFFER, and SUTRACELIST classes

First we define the path to the su file we are going to work with.

```python
su_filespec = r"D:\Data\SEISMICS\Artois201210\shot.su"
```

**supy** provides 4 classes to represent a seismic unix profile:

- **SUSTREAM** is used when the data is opened in stream mode (i.e., a pipe between two processes), which implies that the number of traces is not known up front. This could be useful if you want to develop a stand-alone python script that is run from the shell in combination with other seismic unix programs.
- **SUFILE** is used when the data is read from file, trace by trace. This way of processing is memory-efficient, but not fast. Each trace is read from file sequentially, processed, and then usually needs to be written to another file or object before the next trace is read.
- **SUBUFFER** is used when the data is buffered in one contiguous memory chunk. The data can be read from file or from the standard output of a process or they can be generated by class methods. This requires sufficient memory, but is much faster than the other implementations.
- **SUTRACELIST** is also stored in memory, as a python list of SUTRACE objects. This is less efficient, both in memory and speed, than SUBUFFER. However, it is more flexible, as it allows inserting and shuffling traces.

SUSTREAM objects allow looping over the traces.
SUFILE, SUBUFFER, and SUTRACELIST objects are rewindable, and allow looping, indexing, and slicing of traces.
SUTRACELIST objects allow all methods of a python list (append, insert, pop, remove, ...).
This will be explained later on. SUBUFFER and SUTRACELIST objects allow modifying traces in-place.

All classes provide to_buffer() and to_tracelist() methods to convert to SUBUFFER and SUTRACELIST objects.

All classes can also be operated upon through the generic call_su_prog() method or through dedicated wrappers such as surange(), sugain(), sufilter(), etc.

### SUSTREAM

Let's open the file in stream mode. The input must be a file handle or a pipe. We can loop over a stream, but:

- we don't know the number of traces.
- a stream is not rewindable. Once we have passed over the last trace, we cannot go back to a previous one.
- we can't index, i.e. we can't get a particular trace without looping over the traces before.

```python
import subprocess
ifd = open(su_filespec, "rb")
proc = subprocess.Popen("suwind", stdin=ifd, stdout=subprocess.PIPE)
line = supy.SUSTREAM(proc.stdout)
for tr in line:
    pass
print tr
line.rewind()
tr = line.getnexttrace()
print tr
ifd.close()
```

### SUFILE

Let's read the line in file mode. The input must be a file handle or a string representing the file path. It is important to realize that if we modify a trace in a loop, the trace needs to be written to a file or pipe before the loop variable is assigned a new trace object, otherwise changes will be lost.

```python
line = supy.SUFILE(su_filespec)
print len(line)
for tr in line:
    pass
print tr
line.rewind()
tr = line.getnexttrace()
print tr
line.close()
```

### SUBUFFER

Now, let's read the line in buffer mode. The input can be:

- a file handle
- a pipe
- a string containing the file path
- a binary string
- a bytearray
- None or an empty string

In contrast to SUSTREAM and SUFILE, traces can be modified persistently when looped over or indexed.

```python
line = supy.SUBUFFER(su_filespec)
print len(line)
for tr in line:
    pass
print tr
line.rewind()
tr = line.getnexttrace()
print tr
```

## Class representing a seismic trace: SUTRACE

We have seen that simple looping over a profile (SUSTREAM, SUFILE, SUBUFFER or SUTRACELIST) yields a trace, i.e. a SUTRACE object. The same applies to indexing (see below).

Seismic Unix traces consist of a trace header and data. Header word values are exposed as class properties, and also as keys:

```python
line = supy.SUBUFFER(su_filespec)
tr = line[0]
print tr.dt
print tr["dt"]
```

Header word values can be set using the same syntax

```python
tr.dt = 500
print tr.dt
```

SUTRACE objects derived from SUBUFFER objects represent the same memory. Hence, if you change the trace, the SUBUFFER object is changed as well.

```python
print line[0].dt
```

The doc() method provides an explanation for a particular header word.

```python
print tr.doc("tracl")
```

Besides simple access to header word values, there are also dedicated methods to get or set certain values. The method get_dt_sec() returns the sampling interval in seconds, either dt multiplied by 1E+6 or d1 (if set and trid is 1):

```python
print tr.get_dt_sec()
```

Methods get_sx(), get_sy(), get_gx(), and get_gy() return source and receiver coordinates in meter or degrees, depending on scalco and counit. There are also equivalent set_() methods.

```python
print tr.sx, tr.sy
print tr.get_sx(), tr.get_sy()
```

Methods related to date and time

```python
print tr.year, tr.day
tr.set_date(datetime.date(2012, 10, 9))
print tr.get_date().isoformat()
print tr.year, tr.day, tr.hour, tr.minute, tr.sec
print tr.get_datetime().isoformat()
```

Finally, there are also methods to obtain header word values affected by scalel, e.g. get_gelev(), get_sdepth(), etc.

```python
tr.gelev = 3000
tr.scalel = -1000
print tr.get_gelev()
```

The trace data can be accessed as a numpy array:

```python
print tr.data
```

All numerical functions in numpy, scipy, and matplotlib can thus be applied to the trace data.

```python
print tr.data.min(), tr.data.max()
print tr.data.argmin(), tr.data.argmax()
print np.abs(tr.data)
print tr.data ** 2
print tr.data.mean()
```

We can also loop with a sliding window

## Looping, indexing, and slicing

We will demonstrate different ways of looping, indexing, and slicing using a SUBUFFER object.

```python
line = supy.SUBUFFER(su_filespec)
```

As we have seen above, SUBUFFER objects can be looped over like any python iterable. Looping yields SUTRACE objects. Modification of a trace will be reflected in the SUBUFFER object.

```python
for i, tr in enumerate(line):
    tr.tracl = i
    if i == 96:
        break
    clear_output()
    print i
    sys.stdout.flush()
```

We can also fetch a particular trace by indexing. Indexing is zero-based, and negative indexes  access traces from the end of the line counting backwards. Indexing is identical to calling the method gettrace().

```python
tr = line[0]
print tr.tracl
tr = line[-1]
print tr.tracl
```

Slicing can occur in two ways: using a regular python slice or using a list or array with arbitrary trace indexes:

```python
traces = line[0:12:2]
print [tr.tracl for tr in traces]
```

```python
traces = line[[1,5,8,14,23]]
print [tr.tracl for tr in traces]
```

Note that slicing results in a new SUBUFFER object. Unlike with indexing, modifying traces in this new object, does not change the original SUBUFFER object. This is actually similar to the behavior of numpy arrays.

```python
traces[0].trid = 2
print traces[0].tracl
print line[0].tracl
```

There are also other ways to loop over or slice SUBUFFER objects:

- over sliding windows
- over gathers

```python
for window in line.iter_windows(hwidth=2, symmetric=True):
    print [tr.tracl for tr in window]
```

```python
su_filespec2 = r"D:\Data\SEISMICS\Artois201210\1.su"
line2 = supy.SUBUFFER(su_filespec2)
```

```python
for shot in line.iter_gathers(key1="fldr", key2="tracf", min=100, max=110):
    print shot[0].fldr, shot.ntr
```

We can also select a particular gather or window directly:

```python
gather = line2.get_gather("fldr", 200, key2="tracf")
print gather[0].fldr, gather.ntr
```

Remind that the objects returned by the above methods are new SUBUFFER objects. Modifying traces from these windows or gathers won't be reflected in the original SUBUFFER object.

```python
gather[0].trid= 2
gather = line2.get_gather("fldr", 200, key2="tracf")
print gather[0].trid
```

If we don't want to loose changes applied during a loop, we need to store the result, either by writing it to file or by copying it into a new SUBUFFER object.

```python
folder = os.path.split(su_filespec)[0]
out_filespec = os.path.join(folder, "out.su")
ofd = open(out_filespec, "wb")
for shot in line2.iter_gathers(key1="fldr", key2="tracf", min=100, max=110):
    for tr in shot:
        tr.data *= 2
    shot.write(ofd)
ofd.close()
line3 = supy.SUBUFFER(out_filespec)
line3.suximage(perc=97)
```

Alternatively, it is possible to modify the original SUBUFFER object in-place, by looping over the gather indexes rather than the gathers:

```python
for tr_indexes in line2.iter_gather_indexes(key1="fldr", key2="tracf", min=100, max=110):
    for tr_index in tr_indexes:
        tr = line2.gettrace(tr_index)
        tr.trid = 2
shot = line2.get_gather("fldr", 100)
print shot[0].trid
```

```python
line_buf = supy.SUBUFFER(su_filespec)
line_file = supy.SUFILE(su_filespec)
%timeit [tr for tr in line_buf]
%timeit [tr for tr in line_file]
del line_file
```

## Interacting with seismic unix programs

SUFILE, SUBUFFER, and SUTRACELIST objects can interface through any seismic unix program using the call_su_prog() method or through dedicated methods like surange(), sufilter(), sugain(), etc. The return value is always a new SUBUFFER object.

```python
#su_filespec = r"D:\Data\SEISMICS\Artois201210\1.su"
#line = supy.SUBUFFER(su_filespec)
cof = line.get_gather("tracf", 3)
cof.sufilter(f=[50,100,300,600], amps=[0.,1.,1.,0.]).sugain(agc=1, wagc=0.2).suximage(perc=97)
```
